```python
from openmm.app import *
from openmm import *
from openmm.unit import *
from sys import stdout
```

```python
from pdbfixer import PDBFixer
# fixer = PDBFixer(filename='/home/jok120/Downloads/1as4.pdb')
fixer = PDBFixer(filename='/net/pulsar/home/koes/jok120/repos/sidechainnet/sidechainnet/'
                          'research/analysis/minimization/4P99_1_A.pdb')
fixer.findMissingResidues()
fixer.findMissingAtoms()
print("Missing atoms", fixer.missingAtoms, fixer.missingResidues)
fixer.addMissingAtoms()
fixer.addMissingHydrogens(7)
# PDBFile.writeFile(fixer.topology, fixer.positions, open('/home/jok120/Downloads/1as4_fixed.pdb', 'w'))
```

```python


# pdb = PDBFile('/home/jok120/Downloads/1as4_fixed.pdb')
# pdb = PDBFile('/net/pulsar/home/koes/jok120/repos/sidechainnet/sidechainnet/research/analysis/minimization/4U6F_51_M5.pdb')
# forcefield = ForceField('amber14-all.xml', 'amber14/tip3pfb.xml')
forcefield = ForceField('amber14/protein.ff15ipq.xml', 'amber14/spce.xml')

system = forcefield.createSystem(fixer.topology, nonbondedMethod=app.NoCutoff,
        nonbondedCutoff=1*nanometer, constraints=HBonds)
integrator = LangevinMiddleIntegrator(300*kelvin, 1/picosecond, 0.004*picoseconds)
simulation = Simulation(fixer.topology, system, integrator)
simulation.context.setPositions(fixer.positions)
simulation.minimizeEnergy()
simulation.reporters.append(PDBReporter('/scr/jok120/mintest/4P99_1_A_openmm.pdb', 1000))
simulation.reporters.append(StateDataReporter(stdout, 1000, step=True,
        potentialEnergy=True, temperature=True))
simulation.step(100000)
```

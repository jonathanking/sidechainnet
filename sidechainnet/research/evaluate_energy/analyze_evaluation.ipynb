```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
RESULTS_FILE = "/home/jok120/sidechainnet/sidechainnet/research/evaluate_energy/result07.csv"
df = pd.read_csv(RESULTS_FILE, index_col=0, header=0, na_values=['None']).dropna()
df
```

# Box Plots

```python
prop_cycle = plt.rcParams['axes.prop_cycle']
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
my_palette = {k: cycle[i] for i, k in enumerate(df.columns)}
my_palette
```

```python
# Plot the results as a box plot, with the energy value on the y axis and the
# structure type on the x axis
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
fig, ax = plt.subplots(figsize=(8, 8))
sns.boxplot(data=df[["native", "minimized", "native_eloss", "minimized_eloss", "af2norelax", "af2relax"]], 
    ax=ax, showfliers=False, palette=my_palette)
ax.set_ylabel("Energy (kJ/mol)")
ax.set_xlabel("Structure type")
ax.set_title("Energy comparison")
# Rotate x labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=25, horizontalalignment='center')
plt.show()

```

# Scatterplots

```python
# Plot the results as a scatterplot
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
fig, ax = plt.subplots(figsize=(8, 8))
sns.scatterplot(x="native", y="minimized", data=df, ax=ax)
ax.set_xlabel("native energy (kcal/mol)")
ax.set_ylabel("minimized energy (kcal/mol)")
ax.set_title("Energy comparison")
# Set the x and y limits to be the same
if ax.get_xlim() > ax.get_ylim():
    ax.set_ylim(ax.get_xlim())
else:
    ax.set_xlim(ax.get_ylim())
# Draw the diagonal line
ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c=".3")
plt.show()
```

```python
# Plot the results as a scatterplot
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
fig, ax = plt.subplots(figsize=(8, 8))
# Show the correlation between the native and af2norelax energies in a scatterplot
sns.scatterplot(data=df, x="native", y="minimized", ax=ax, s=10)
ax.set_xlabel("Native energy (kcal/mol)")
ax.set_ylabel("minimized energy (kcal/mol)")
ax.set_title("Energy comparison")
# Set the x and y limits to be the same
if ax.get_xlim() > ax.get_ylim():
    ax.set_ylim(ax.get_xlim())
else:
    ax.set_xlim(ax.get_ylim())

ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c=".3")
plt.show()
```
